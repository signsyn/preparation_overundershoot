```python
from scipy.io import wavfile
import pyworld as pw
import pysptk as ps
import librosa
import numpy as np
import matplotlib.pyplot as plt
import IPython.display as display
import math
import seaborn
import soundfile

WAV_FILE1 = '../../share/RIKIYA/あ.wav'  # １音目
WAV_FILE2 = '../../share/RIKIYA/お.wav'  # ２音目

data1, fs1 = soundfile.read(WAV_FILE1)
data1 = data1.astype(np.float)  # WORLDはfloat前提

data2, fs2 = soundfile.read(WAV_FILE2)
data2 = data2.astype(np.float)  # WORLDはfloat前提

pitch1 = 60  # C4
pitch2 = 69  # A4

def extract(data, fs):  # f0, メルケプ, bapを抽出する関数
    _f0, t = pw.dio(data, fs)  # 基本周波数の抽出
    f0 = pw.stonemask(data, _f0, t, fs)  # 基本周波数の修正
    sp = pw.cheaptrick(data, f0, t, fs)  # スペクトル包絡の抽出
    order = 60
    alpha = 0.544
    mcep = ps.conversion.sp2mc(sp, order, alpha)  # メルケプストラム
    ap = pw.d4c(data, f0, t, fs)  # 非周期性指標の抽出
    bap = np.zeros((len(t), 5))
    bap[:, 0] = np.average(ap[:, : 128])
    bap[:, 1] = np.average(ap[:, 128: 256])
    bap[:, 2] = np.average(ap[:, 256: 512])
    bap[:, 3] = np.average(ap[:, 512: 768])
    bap[:, 4] = np.average(ap[:, 768: 1025])  # bap
    return f0, mcep, bap

f01, mcep1, bap1 = extract(data1, fs1)
f02, mcep2, bap2 = extract(data2, fs2)

def linear(data1,data2,point):  # 関数leng内で使用
    data = (data2 - data1) * point + data1
    if(point < 0 or point > 1):
        sys.stderr.write("please input 0 < point < 1\n")
    return data

def leng(data,f0,mcep,bap,time,ran):  # 長さを調節する関数
    y = 201.62  # 201.62は1秒のf0等のフレーム数
    frame = math.floor(time * y + ran)
    x = (len(f0) - 1) / frame
    
    newf0 = np.zeros(math.floor(frame))
    newmcep = np.zeros((math.floor(frame), mcep.shape[1]))
    newbap = np.zeros((math.floor(frame), bap.shape[1]))
    
    ind = 0
    
    for k in range(len(newf0)):
        f,d = math.modf(k * x)
        
        for l in range(mcep.shape[1]):
            newmcep[k,l] = linear(mcep[math.floor(ind),l], mcep[math.ceil(ind),l], f)
        for l in range(bap.shape[1]):
            newbap[k,l] = linear(bap[math.floor(ind),l], bap[math.ceil(ind),l], f)
            
        newf0[k] = linear(f0[math.floor(ind)], f0[math.ceil(ind)], f)
        
        ind = ind + x
        
    return newf0,newmcep,newbap

ran = 20  # フレーム
time1 = 1  # 秒
time2 = 1  # 秒

f01, mcep1, bap1 = leng(data1, f01, mcep1, bap1, time1, ran)
f02, mcep2, bap2 = leng(data2, f02, mcep2, bap2, time2, ran)

def makePitch(f0, pitch):  # constant F0
    targetF0 = 440 * pow(2, (pitch - 69) / 12)
    newF0 = np.vectorize(lambda x: 0.0 if x == 0 else targetF0)(f0)
    return newF0.astype(np.double)

f01 = makePitch(f01, pitch1)
f02 = makePitch(f02, pitch2)

# 可視化
def audiomake(f0, mcep, bap, fs):  # 音を生成
    alpha = 0.55
    fftlen = 2048
    sp = ps.conversion.mc2sp(mcep, alpha, fftlen)
    ap = np.zeros((len(f0), 1025))
    for k in range(128):
        ap[:, k] = bap[:, 0]
    for k in range(128):
        ap[:, 128+k] = bap[:, 1]
    for k in range(256):
        ap[:, 256+k] = bap[:, 2]
    for k in range(256):
        ap[:, 512+k] = bap[:, 3]
    for k in range(257):
        ap[:, 768+k] = bap[:, 4]  # bap
    synthesized = pw.synthesize(f0, sp, ap, fs)
    return synthesized

syn1 = audiomake(f01, mcep1, bap1, fs1)
syn2 = audiomake(f02, mcep2, bap2, fs2)

print('あ１')
display.display(
    display.Audio(syn1, rate=fs1)
)

plt.plot(data1, label="Raw Data")
plt.legend(fontsize=10)
plt.show();

plt.plot(f01, linewidth=3, color="green", label="F0 contour")  # １音目のf0のプロット
plt.legend(fontsize=10)
plt.show();

print('お２')
display.display(
    display.Audio(syn2, rate=fs2)
)

plt.plot(data2, label="Raw Data")
plt.legend(fontsize=10)
plt.show();

plt.plot(f02, linewidth=3, color="green", label="F0 contour")  # ２音目のf0のプロット
plt.legend(fontsize=10)
plt.show();
```

```python
def hokan1(f01,f02,mcep1,mcep2,bap1,bap2,ran,targetF01,targetF02):  # 線形補間
    newf0 = modifyF0(f01,f02,ran,targetF01,targetF02)
    newmcep = np.zeros((mcep1.shape[0]+mcep2.shape[0]-ran*2,mcep1.shape[1]))
    newbap = np.zeros((bap1.shape[0]+bap2.shape[0]-ran*2,bap1.shape[1]))
    for k in range(mcep1.shape[1]):
        newmcep[:,k] = modify(mcep1[:,k],mcep2[:,k],ran)
    for k in range(bap1.shape[1]):
        newbap[:,k] = modify(bap1[:,k],bap2[:,k],ran)
    return newf0,newmcep,newbap


def modify(data1,data2,ran):
    newdata = np.zeros(len(data1)+len(data2)-ran*2)
    for k in range(len(newdata)):
        if k < len(data1)-ran*2:
            newdata[k] = data1[k]
        elif k < len(data1):
            newdata[k] = data1[k]*((len(data1)-k-1)/(ran*2))+data2[k+ran*2-len(data1)]*((k+ran*2-len(data1)+1)/(ran*2))
        else:
            newdata[k] = data2[k+ran*2-len(data1)]
    return newdata

def modifyF0(data1,data2,ran,targetF01,targetF02):  # 補間時のf0をtargetF0を用いて計算する
    newdata = np.zeros(len(data1)+len(data2)-ran*2)
    for k in range(len(newdata)):
        if k < len(data1)-ran*2:
            newdata[k] = data1[k]
        elif k < len(data1):
            newdata[k] = targetF01*((len(data1)-k-1)/(ran*2))+targetF02*((k+ran*2-len(data1)+1)/(ran*2))
        else:
            newdata[k] = data2[k+ran*2-len(data1)]
    return newdata

targetF01 = 440 * pow(2, (pitch1 - 69) / 12)
targetF02 = 440 * pow(2, (pitch2 - 69) / 12)

ran = 10
f04, mcep4, bap4 = hokan1(f01,f02,mcep1,mcep2,bap1,bap2,ran,targetF01,targetF02)
syn4 = audiomake(f04, mcep4, bap4, fs1)

print('線形補間')
display.display(
    display.Audio(syn4, rate=fs1)
)

plt.plot(f04, linewidth=3, color="green", label="F0 contour")  # f0のプロット
plt.legend(fontsize=10)
plt.show();

plt.plot(f04, linewidth=3, color="green", label="F0 contour")
plt.legend(fontsize=10)
plt.xlim(150, 250)
plt.ylim(200, 550)
plt.show();
```

```python
# 三次関数でプレパレーション，オーバーシュート・アンダーシュート

def pouconnect(f01,f02,mcep1,mcep2,bap1,bap2,p_ran1,p_ran2,p_ran3,ou_ran1,ou_ran2,ou_ran3,targetF01,targetF02):  # プレパレーションとオーバーシュート・アンダーシュート両方
    p_f05, mcep5, bap5 = preparation(f01,f02,mcep1,mcep2,bap1,bap2,p_ran1,p_ran2,p_ran3,targetF01,targetF02)
    ou_f05, _, _ = overundershoot(f01,f02,mcep1,mcep2,bap1,bap2,ou_ran1,ou_ran2,ou_ran3,targetF01,targetF02)
    f05 = np.concatenate((p_f05[: len(f01)-p_ran2],ou_f05[len(f01)-p_ran2: len(ou_f05)]))
    return  f05, mcep5, bap5

def preparation(f01,f02,mcep1,mcep2,bap1,bap2,ran1,ran2,ran3,targetF01,targetF02):  # プレパレーション
    l = (targetF02 - targetF01) / (ran2 * 2)
    x1 = len(f01) - ran2 - ran3
    y1 = targetF01
    x2 = len(f01) - ran2 - ran1
    y2 = targetF01 + l * (ran2 - ran1)
    C1 = matrix(x1,y1,x2,y2,l)
    f05, mcep5, bap5 = hokan1(f01,f02,mcep1,mcep2,bap1,bap2,ran2,targetF01,targetF02)
    for k in range(ran3 - ran1):
        t1 = len(f01)-ran2-ran3+k
        f05[t1-1] = C1[0,0]*(t1)**3 + C1[1,0]*(t1)**2 + C1[2,0]*t1 + C1[3,0]  # t1-1でうまくいったのはなぜ？
    return f05, mcep5, bap5

def overundershoot(f01,f02,mcep1,mcep2,bap1,bap2,ran1,ran2,ran3,targetF01,targetF02):  # オーバーシュート・アンダーシュート
    l = (targetF02 - targetF01) / (ran2 * 2)
    x1 = len(f01) - ran2 - ran3
    y1 = targetF01
    x2 = len(f01) - ran2 - ran1
    y2 = targetF01 + l * (ran2 - ran1)
    C2 = matrix(x1,y1,x2,y2,l)
    f05, mcep5, bap5 = hokan1(f01,f02,mcep1,mcep2,bap1,bap2,ran2,targetF01,targetF02)
    for k in range(ran3 - ran1):
        t1 = len(f01)-ran2-ran3+k
        t2 = len(f01)-ran2+ran3-k
        f05[t2-1] = targetF01 + targetF02 - (C2[0,0]*(t1)**3 + C2[1,0]*(t1)**2 + C2[2,0]*t1 + C2[3,0])  # t2-1でうまくいったのはなぜ？
    return f05, mcep5, bap5

def matrix(x1,y1,x2,y2,l):  # 三次関数の導出
    A = np.array([[x1**3,x1**2,x1,1],[x2**3,x2**2,x2,1],[3*x1**2,2*x1,1,0],[3*x2**2,2*x2,1,0]])
    B = np.array([[y1],[y2],[0],[l]])
    inv_A = np.linalg.inv(A)
    C = np.dot(inv_A, B)
    return C

# 条件 ran3 > ran2 > ran1

p_ran1 = 5
p_ran2 = 10  # 線形補間のranと同じにする
p_ran3 = 35

ou_ran1 = 5
ou_ran2 = 10  # 線形補間のranと同じにする
ou_ran3 = 35

f05, mcep5, bap5 = pouconnect(f01,f02,mcep1,mcep2,bap1,bap2,p_ran1,p_ran2,p_ran3,ou_ran1,ou_ran2,ou_ran3,targetF01,targetF02)

syn5 = audiomake(f05, mcep5, bap5, fs1)

print('三次関数＋線形補間')
display.display(
    display.Audio(syn5, rate=fs1)
)

plt.plot(f05, linewidth=3, color="green", label="F0 contour")  # f0のプロット
plt.legend(fontsize=10)
plt.show();

plt.plot(f05, linewidth=3, color="green", label="F0 contour")
plt.legend(fontsize=10)
plt.xlim(150, 250)
plt.ylim(200, 550)
plt.show();
```
